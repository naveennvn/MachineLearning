```python
Model_Name = []
Engine_Type = []
Price = []


for i in mobile_tag1:
    tag_text = i.text
    ######### model name##########
    model = re.findall(r'\b[GE].*',tag_text)
    #print(model)
    Model_Name.append(''.join(model))
    #print(Model_Name)
    
    ########## Engine Type############
    engine = re.findall(r'\b[PC].*',tag_text)
    Engine_Type.append(''.join(engine))
    #print(Engine_Type)
    
    
    ######## Price###########
    price = re.findall(r'\d+\W\d+\W\d+',tag_text)
    Price.append(''.join(price))
    print(price)
```
